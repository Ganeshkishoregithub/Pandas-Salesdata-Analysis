```python
all_data=pd.read_csv("all_data.csv")
all_data
```

```python
#Adding a column Month
all_data['Month']=all_data['Order Date'].str[0:2]
all_data

#As we have converted Month to str we need to change it to int. Befor that we could see Nan in Month we need to drop them to proceed
```

```python
#Dropping Nan
nan_data=all_data[all_data.isna().any(axis=1)]
nan_data
```

```python
all_data=all_data.dropna(how='all')
all_data
```

```python
#Converting Month to int
all_data['Month']=all_data['Month'].astype('int32')

#We could see some or so we need to remove them before converting
```

```python

or_data=all_data[all_data['Order Date'].str[0:2]=='Or']
or_data
```

```python
all_data=all_data[all_data['Month']!='Or']
all_data
```

```python
#Now converting it to int
all_data['Month']=all_data['Month'].astype('int32')
all_data
```

```python
#Q1- What is the best month for sales? how much earned that month

# for this we need to multiply Quantity Ordered and Price Each
```

```python
#Before that we need to change Quantity Ordered and Price Each to numberic type
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered']) #To int
all_data['Price Each']=pd.to_numeric(all_data['Price Each']) # to float
```

```python
all_data['Sales']= all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
#Q1- What is the best month for sales? how much earned that month
results=all_data.groupby('Month').sum()
```

```python
#To display as bar chart
import matplotlib.pyplot as plt
months =range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.xlabel('Months Number')
plt.ylabel('Sales in USD($)')
plt.plot
```

```python
#What product are most often sold together?
```

```python
#For this we need to combine all the prducts which has the same  order id
sameorder=all_data[all_data['Order ID'].duplicated(keep=False)]
sameorder
```

```python
#Now the same orderis repeated to avoid that
sameorder['Grouped']=sameorder.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
sameorder
```

```python
#Now need to remove duplicateds
sameorder=sameorder[['Order ID', 'Grouped']].drop_duplicates()
sameorder.head(20)
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in sameorder['Grouped']:
    row_list=row.split(',')
    #print(row_list)
    count.update(Counter(combinations(row_list,2)))
print(count)
```

```python
#for proper formatting
for key, value in count.most_common(10):
    print(key,value)
```

```python
product_group=all_data.groupby('Product')
product_group.sum()


```

```python
quantity_ordered= product_group.sum()['Quantity Ordered']
products=[product for product, df in product_group]
plt.bar(products,quantity_ordered)
plt.xticks(products,rotation='vertical',size=7)
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
#Now we need to find the reason why the products are sold highly- however the reason should be price
prices=all_data.groupby('Product').mean()['Price Each']
print(prices)
```

```python
#Now ploting second y label 
fig , ax1 =plt.subplots()
ax2= ax1.twinx()
ax1.bar(products, quantity_ordered, color='r')
ax2.plot(products, prices, color='y')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='r')
ax2.set_xlabel('Product Name')
ax2.set_ylabel('Price in USD',color='y')
ax1.set_xticklabels(products,rotation='vertical',size=8)
plt.show()
```
